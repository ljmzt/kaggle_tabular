```python
data = data[colF4]
sc = MinMaxScaler(feature_range=(-1,1))
X_norm = sc.fit_transform(data.values)
means = np.nanmean(X_norm, axis=0)
means
```

```python
X_todo = X_norm[np.sum(np.isnan(X_norm),axis=1)>0].copy()
X_todo.shape
```

```python
class Model(nn.Module):
    def __init__(self, input_dim=14, hidden_dim=128):
        super().__init__()
        self.fc1 = nn.Linear(input_dim, hidden_dim*4)
        self.bn1 = nn.BatchNorm1d(hidden_dim*4)
        self.fc2 = nn.Linear(hidden_dim*4, hidden_dim*2)
        self.bn2 = nn.BatchNorm1d(hidden_dim*2)
        self.fc3 = nn.Linear(hidden_dim*2, hidden_dim)
        self.bn3 = nn.BatchNorm1d(hidden_dim)
        self.fc4 = nn.Linear(hidden_dim, 1)
        self.activation = nn.ReLU()
    def forward(self, x):
        x = self.bn1(self.activation(self.fc1(x)))
        x = self.bn2(self.activation(self.fc2(x)))
        x = self.bn3(self.activation(self.fc3(x)))
        return nn.Tanh()(self.fc4(x))
```

```python
models = {}
for icol in range(15):
    model = Model()
    model.load_state_dict(torch.load(f'model_0_{icol}.pickle')['model'])
    model.eval()
    models[(0,icol)] = model
    
    model = Model()
    model.load_state_dict(torch.load(f'model_1_{icol}.pickle')['model'])
    model.eval()
    models[(1,icol)] = model
    
    model = Model()
    model.load_state_dict(torch.load(f'model_2_{icol}.pickle')['model'])
    model.eval()
    models[(2,icol)] = model
```

```python
class ModelOneShot(nn.Module):
    def __init__(self, input_dim=15, hidden_dim=256):
        super().__init__()
        self.fc1 = nn.Linear(input_dim, hidden_dim*4)
        self.bn1 = nn.BatchNorm1d(hidden_dim*4)
        self.fc2 = nn.Linear(hidden_dim*4, hidden_dim*4)
        self.bn2 = nn.BatchNorm1d(hidden_dim*4)
        self.fc3 = nn.Linear(hidden_dim*4, hidden_dim*2)
        self.bn3 = nn.BatchNorm1d(hidden_dim*2)
        self.fc4 = nn.Linear(hidden_dim*2, hidden_dim)
        self.bn4 = nn.BatchNorm1d(hidden_dim)
        self.fc5 = nn.Linear(hidden_dim, input_dim)
        self.activation = nn.ReLU()
        self.dropout = nn.Dropout(p=0.2)
    def forward(self, x):
        x = self.bn1(self.activation(self.fc1(x)))
        x = self.dropout(x)
        x = self.bn2(self.activation(self.fc2(x)))
        x = self.dropout(x)
        x = self.bn3(self.activation(self.fc3(x)))
        x = self.dropout(x)
        x = self.bn4(self.activation(self.fc4(x)))
        return nn.Tanh()(self.fc5(x))
```

```python
idx0 = np.arange(15)
for i in range(240732):
    if (i % 1000 ==0):  print(i)
    x_todo = X_todo[i].copy()
    idx = idx0[np.isnan(x_todo)]
    nfill = len(idx)
        
    # process 3 missing to see
    if nfill <= 3:
        # first fill with -1
        for col in idx:
            x_todo[col] = -1
        
        for col in idx:
            model = models[(nfill-1,col)]
            x = np.hstack((x_todo[:col],x_todo[col+1:]))
            x = torch.Tensor(x)[None,:]
            y = model(x).item()
            x_todo[col] = y
    
    else:
        for col in idx:
            x_todo[col] = means[col]
    
    # reset the X_todo
    for col in idx:
        X_todo[i,col] = x_todo[col]
    
```

```python
model = ModelOneShot()
model.load_state_dict(torch.load('model_oneshort.pickle')['model'])
model.eval()
for i in range(240732):
    if (i % 1000 ==0):  print(i)
    x_todo = X_todo[i].copy()
    idx = idx0[np.isnan(x_todo)]
    nfill = len(idx)
        
    # process 3 missing to see
    if nfill <= 3:
        # first fill with -1
        x_todo[idx] = -1
        x = torch.Tensor(x_todo)[None,:]
        predict = model(x)[0]
        for col in idx:
            x_todo[col] = predict[col]
    
    else:
        for col in idx:
            x_todo[col] = means[col]
    
    # reset the X_todo
    for col in idx:
        X_todo[i,col] = x_todo[col]
    
```
