```python
def search(X0_norm, X1_norm, n_series=1000, n_win=10, template_len=15):
    results = []
    ids = np.random.choice(range(len(X1_norm)), n_series)
    best1, best2 = -999, -999
    for idx in ids:
        for i in np.random.choice(range(len(X1_norm[0])-template_len), n_win):
            template = X1_norm[idx, i:i+template_len].copy()
            template -= np.mean(template)
            template /= np.sqrt(np.sum(np.power(template,2)))
            
            X1_X1 = correlate(template[np.newaxis,:], X1_norm, mode='valid')[::-1]
            X1_X1_max = np.max(X1_X1, axis=1)
            X1_X1_max_median = np.median(X1_X1_max)
            X1_X1_max_std = np.std(X1_X1_max)
            
            X0_X0 = correlate(template[np.newaxis,:], X0_norm, mode='valid')[::-1]
            X0_X0_max = np.max(X0_X0, axis=1)
            X0_X0_max_median = np.median(X0_X0_max)
            X0_X0_max_std = np.std(X0_X0_max)
            
            results.append([idx,i, X1_X1_max_median, X1_X1_max_std,
                            X0_X0_max_median, X0_X0_max_std])
            if (X1_X1_max_median - X0_X0_max_median > best1):
                best1 = X1_X1_max_median - X0_X0_max_median
                best_template1 = template.copy()
            if (X1_X1_max_median - X0_X0_max_median - X1_X1_max_std - X0_X0_max_std > best2):
                best2 = X1_X1_max_median - X0_X0_max_median - X1_X1_max_std - X0_X0_max_std
                best_template2 = template.copy()

    return np.array(results), best_template1, best_template2
```

```python
def get_template(X1_norm, results, kk, template_len=15):
    templates = []
    for k in kk:
        idx, i = np.int32(results[k,0]), np.int32(results[k,1])
        template = X1_norm[idx, i:i+template_len].copy()
        template -= np.mean(template)
        template /= np.sqrt(np.sum(np.power(template,2)))
        templates.append(template)
    return np.array(templates)
```
