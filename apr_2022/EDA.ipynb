```python
def distance(seq1, seq2):
    mean_tmp = np.delete(mean,2,axis=1)
    std_tmp = np.delete(std,2,axis=1)
    a = series.loc[seq1].values
    a = np.delete(a,2,axis=1)  #sensor 2 seems broken
    a = ((a-mean_tmp)/std_tmp).flatten()
    b = series.loc[seq2].values
    b = np.delete(b,2,axis=1)
    b = ((b-mean_tmp)/std_tmp).flatten()
    return np.dot(a,b) / np.sqrt(np.dot(a,a)*np.dot(b,b))
```

```python
dd = np.zeros((len(seqs),len(seqs)))
for id1,seq1 in enumerate(seqs):
    for id2,seq2 in enumerate(seqs):
        if (seq1 == seq2):  continue
        dd[id1,id2] = distance(seq1,seq2)
sns.heatmap(dd,vmin=-0.2,vmax=0.2)
```

```python
def distance2(seq1, seq2):
    mean_tmp = np.delete(mean,2,axis=1)
    std_tmp = np.delete(std,2,axis=1)
    a = series.loc[seq1].values
    a = np.delete(a,2,axis=1)  #sensor 2 seems broken
    a = (a - mean_tmp) / std_tmp
    a = np.corrcoef(a,rowvar=False).flatten()
    
    b = series.loc[seq2].values
    b = np.delete(b,2,axis=1)  #sensor 2 seems broken
    b = (b - mean_tmp) / std_tmp
    b = np.corrcoef(b,rowvar=False).flatten()
    
    return np.dot(a,b) / np.sqrt(np.dot(a,a)*np.dot(b,b))
```

```python
dd = np.zeros((len(seqs),len(seqs)))
for id1,seq1 in enumerate(seqs):
    for id2,seq2 in enumerate(seqs):
        dd[id1,id2] = distance2(seq1,seq2)
sns.heatmap(dd)
```
