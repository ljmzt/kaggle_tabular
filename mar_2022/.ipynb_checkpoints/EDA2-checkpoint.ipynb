```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
train_dat = pd.read_csv('train.csv', index_col='row_id')
# train_dat = train_dat.loc[0:5000]
train_dat.head(10)
```

```python
mapper = {}
ii = 0
for h in range(24):
    for mm in ['00','20','40']:
        hh = '{0:02d}'.format(h)
        mapper[hh+':'+mm] = ii
        ii += 1
time_mapper = mapper.copy()
time_mapper
```

```python
mapper = {}
ii = 0
for h in range(24):
    for mm in ['00','20','40']:
        hh = '{0:02d}'.format(h)
        mapper[hh+':'+mm] = ii
        ii += 1
train_dat['time_stamp'] = train_dat['time'].apply(lambda x:mapper[x.split()[1][:5]])
# train_dat['time'] = pd.to_datetime(train_dat['time'])
train_dat.head()
```

```python
train_dat['unique'] = train_dat['x'].astype(str) + train_dat['y'].astype(str) + train_dat['direction']
train_dat['day'] = pd.to_datetime(train_dat['time']).dt.weekday

tmp = train_dat.groupby(['unique','day','time_stamp']).agg({'congestion':np.median})
mapper = tmp.to_dict()['congestion']
train_dat['median'] = train_dat.apply(lambda x: mapper[(x['unique'], x['day'], x['time_stamp'])], axis=1)
# train_dat
```

```python
median_mapper = mapper.copy()
median_mapper
```

```python
time_mapper
```

```python
congestion_table_withtime['time_stamp'] = \
   pd.Series(congestion_table_withtime.index).apply(lambda x:time_mapper[x.split()[1][:5]]).values
```

```python
congestion_table_withtime['day'] = \
   pd.to_datetime(pd.Series(congestion_table_withtime.index)).dt.weekday.values
congestion_table_withtime
```

```python
def assemble(col):
    cols = [col for col in congestion_table_withtime.columns.to_list() if '-' in col]
    cols.append('time_stamp')
    cols.append('day')
    train = congestion_table_withtime[cols].copy()  #the copy is needed
    train['median'] = train.apply(\
                        lambda x: median_mapper[(col, x['day'], x['time_stamp'])], axis=1).values
    train['target'] = congestion_table_withtime[col].values
    return train
```

```python
congestion_table_withtime
```

```python
train = assemble('00NB')
train
```

```python
train.corrwith(train['target']).sort_values()
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
regr = RandomForestRegressor(min_samples_leaf=3, criterion='absolute_error', n_jobs=2)
```

```python
X, y = train.drop(columns=['target']).values, train['target'].values
```

```python
regr.fit(X,y)
```

```python
regr.score(X,y)
```
